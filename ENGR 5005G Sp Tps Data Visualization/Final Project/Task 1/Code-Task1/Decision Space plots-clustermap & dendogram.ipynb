```python
#Import libraries
import numpy
#Remove the column labels and upload the decision space
filename = 'test2.csv'
raw_data = open(filename, 'rt')
data = numpy.loadtxt(raw_data, delimiter=",", skiprows=1)
print(data.shape)
```

```python
import pandas as pd, seaborn as sns
import scipy.spatial as sp, scipy.cluster.hierarchy as hc

sns.set(font="monospace")

sns.set(font_scale=1.9)

X = data
DF = pd.DataFrame(X, index = ["data_%d" % (i) for i in range(X.shape[0])], columns = columns)

#DF = pd.DataFrame(X, columns = columns)

DF_corr = DF.T.corr()
DF_dism = 1 - DF_corr   # distance matrix
linkage = hc.linkage(sp.distance.squareform(DF_dism), method='average')
sns.clustermap(DF_dism, row_linkage=linkage, col_linkage=linkage, yticklabels=True, xticklabels=True, figsize=(40, 40))
```
