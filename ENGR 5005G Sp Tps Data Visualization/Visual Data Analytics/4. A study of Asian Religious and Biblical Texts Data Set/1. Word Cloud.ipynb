Every part of the world follow different sacred texts to learn and preach about their religion.
The following dataset was collected for different religions which is followed in different countries:
1. Hinduism (India): Yogasutras, Upanishads
2. Buddhism (Tibet): Four Noble Truth of Buddhism
3. Taoism (China): Tao Te Ching
4. Christianity (Central Asia/America): Book of Proverb, Book of Ecclesiastes, Book of Ecclesiasticus, Book of Wisdom

The attributes are just the words from the bag of words preprocessing of the mini-corpus made up of the 8 religious books considered in this study. There are 8265 words used. The dataset has raw texts along with  pre-processed **Document Term Matrices (DTM)**.

```python
#to install wordcloud - execute "pip install wordcloud" from Anaconda prompt

#read the course descriptions
with open("data1.txt", 'r') as fh:  
    filedata = fh.read()

#check contents
print("File data sample : ", filedata[:200])

# Create stopword list:
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

#Generate wordcloud data
wordcloud = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(filedata)
```

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

```python
#Add more words to ignore 
#stopwords.update(["to","an", "others", "you","that", "know"])

#Redo stop words. Limit number of words
wordcloud = WordCloud(stopwords=stopwords, max_words=10, \
                      background_color="azure").generate(filedata)

mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```

## 2. Masked word cloud to a image of book

```python
from PIL import Image
book_mask = np.array(Image.open("1.jpg"))
#book_mask
```

```python
# Create a word cloud image
wc = WordCloud(background_color="white", max_words=1000, mask=book_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(filedata)

# store to file
wc.to_file("1.png")

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
